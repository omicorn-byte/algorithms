# Karatsuba Multiplication
Assumptions :
Both number have equal and even digits for simplification

```python
def multiply(x,y):
    p=str(x)
    q=str(y)
    m=len(p)
    n=len(q)
    #print(n//2)
    if (m==n) and m%2==0 and m>=2:
        a=int(p[:m//2])
        b=int(p[m//2:])
        c=int(q[:n//2])
        d=int(q[n//2:])
        ac=multiply(a,c)
        bd=multiply(b,d)
        cc=multiply(a+b,c+d)
        gauss=cc-ac-bd
        #print('y')
        return (10**m)*ac+(10**(m//2))*gauss+bd
    else: 
        #print('n')
        return x*y
```

##  Merge Sort 

```python
def Merge(a,b):
    i=len(a)-1
    j=len(b)-1
    p=0
    q=0
    c=[]
    for _ in range(i+j+2):
        if p<=i and q<=j:
            if a[p]<=b[q]:
                c.append(a[p])
                p+=1
            elif a[p]>b[q]:
                c.append(b[q])
                q+=1
        elif p>i:
            c+=b[q:]
            break
        elif q>j:
            c+=a[p:]
            break
    return c       
```

```python
def Sort(lst):
    i=len(lst)
    if i>1:
        a=Sort(lst[:i//2])
        b=Sort(lst[i//2:])
        return Merge(a,b)
    else:
        return lst
```

```python
%timeit range(100)
Sort([5,4,-2,-6,8,9,9,900])
```

```python
%timeit range(100)
sorted([5,4,-2,-6,8,9,9,900])
```
